<font size="4">    <b>Clean feature definitions</b>

```python
def rms(values):
    return np.sqrt(sum(values**2)/len(values))

def mini(values):
    return np.min(values)

def maxi(values):
    return np.max(values)

def mav(values):
    return np.mean(abs(values))

def var(values):
    return np.var(values)
```

<font size="4">    <b>Not clean feature extraction</b>

```python
#Waveform length
wl_1 = np.zeros((len(data_emg['chan_1'])//win_emg))
b = np.zeros((win_emg))
wl_2 = np.zeros((len(data_emg['chan_2'])//win_emg))
c = np.zeros((win_emg))

for j in range(len(data_emg['chan_1'])//win_emg):
    wl_1[j] = np.sum(b)
    wl_2[j] = np.sum(c)
    
    for i in range(win_emg):
        b[i] = abs(data_emg['chan_1'][(i+1)+j] - data_emg['chan_1'][i+j])
        c[i] = abs(data_emg['chan_2'][(i+1)+j] - data_emg['chan_2'][i+j])
        
#Mean absolute value slope
d_1 = data_emg.groupby(win_emg_scaled)['chan_1'].apply(mav)
mavs_1 = np.zeros((len(data_emg['chan_1'])//win_emg))
d_2 = data_emg.groupby(win_emg_scaled)['chan_2'].apply(mav)
mavs_2 = np.zeros((len(data_emg['chan_2'])//win_emg))

for i in range((len(data_emg['chan_1'])//win_emg)-1):
    mavs_1[i] = d_1[i+1] - d_1[i]
    mavs_2[i] = d_2[i+1] - d_2[i]   
```

```python
#Zero crossing                 Verify if it works on real data
z_1 = np.zeros((len(data_emg['chan_1'])//win_emg))
z_2 = np.zeros((len(data_emg['chan_1'])//win_emg))

for j in range(len(data_emg['chan_1'])//win_emg):
    
    for i in range(win_emg):
        if data_emg['chan_1'][i+j] > 0 and data_emg['chan_1'][(i+1)+j] < 0 and abs(data_emg['chan_1'][i+j]-data_emg['chan_1'][(i+1)+j]) > 1:
            z_1[j] += 1
        elif data_emg['chan_1'][i+j] < 0 and data_emg['chan_1'][(i+1)+j] > 0 and abs(data_emg['chan_1'][i+j]-data_emg['chan_1'][(i+1)+j]) > 1:
            z_1[j] += 1
        else:
            z_1[j] += 0

        if data_emg['chan_2'][i+j]>0 and data_emg['chan_2'][(i+1)+j]<0 and abs(data_emg['chan_2'][i+j]-data_emg['chan_2'][(i+1)+j])>1:
            z_2[j] += 1
        elif data_emg['chan_2'][i+j]<0 and data_emg['chan_2'][(i+1)+j]>0 and abs(data_emg['chan_2'][i+j]-data_emg['chan_2'][(i+1)+j])>1:
            z_2[j] += 1
        else:
            z_2[j] += 0
```

```python
#Slope sign change
ss_1 = np.zeros((len(data_emg['chan_1'])//win_emg))
ss_2 = np.zeros((len(data_emg['chan_1'])//win_emg))

for j in range(len(data_emg['chan_1'])//win_emg):
    
    for i in range(1,win_emg):
        if data_emg['chan_1'][i+j] > data_emg['chan_1'][(i-1)+j] and data_emg['chan_1'][i+j] > data_emg['chan_1'][(i+1)+j]:
            ss_1 += 1
        elif data_emg['chan_1'][i+j] < data_emg['chan_1'][(i-1)+j] and data_emg['chan_1'][i+j] < data_emg['chan_1'][(i+1)+j]:
            ss_1 += 1
        else:
            ss_1 += 0
        
        if data_emg['chan_2'][i+j] > data_emg['chan_2'][(i-1)+j] and data_emg['chan_2'][i+j] > data_emg['chan_2'][(i+1)+j]:
            ss_2 += 1
        elif data_emg['chan_2'][i+j] < data_emg['chan_2'][(i-1)+j] and data_emg['chan_2'][i+j] < data_emg['chan_2'][(i+1)+j]:
            ss_2 += 1
        else:
            ss_2 += 0
```

EMG: We want to extract X features (5 ? See if selection) at each window having a size of 100 samples, so 10 times per epoch or per seconds.

```python
# Write the features list we want to use for the EMG data.
feat_list = [rms, mini, maxi, mav, var]                                        #list for apply groupby
nfeat_list = ['rms', 'mini', 'maxi', 'mav', 'var', 'wl', 'mavs', 'zc', 'ssc']  #list for col name

muscle_1 =[None] * len(feat_list)
muscle_2 =[None] * len(feat_list)

# Exctract the features at each window size for each channel for each epoch and put them in a DataFrame

#EEG 1st DataFrame
win_eeg_mean = data_eeg.groupby(win_eeg_scaled).mean() 

#EMG 2nd DataFrame
for i in range(len(feat_list)):
    muscle_1[i] = data_emg.groupby(win_emg_scaled)['chan_1'].apply(feat_list[i])
    muscle_2[i] = data_emg.groupby(win_emg_scaled)['chan_2'].apply(feat_list[i])

muscle_1.append(wl_1) 
muscle_2.append(wl_2)
muscle_1.append(mavs_1) 
muscle_2.append(mavs_2)
muscle_1.append(z_1) 
muscle_2.append(z_2)
muscle_1.append(ss_1) 
muscle_2.append(ss_2)

m1 = pd.DataFrame(muscle_1, index=nfeat_list).transpose()
m2 = pd.DataFrame(muscle_2, index=nfeat_list).transpose()
mk = {'Biceps':m1, 'Deltoïd':m2}
muscles = pd.concat([m1,m2], axis=1, keys=mk.keys())
c_labels = data_emg.iloc[:,:3].groupby(win_emg_scaled).mean() # Labels for the EMG DataFrame

win_emg_mean = pd.concat([c_labels, muscles], axis=1)
```

<font size="3">    <b>EEG Feature extraction on sliding windows</b>

```python
def slid_win(eeg_chans):
    '''
    Step 1: Compute the mean every feature_size samples for all channels
    Step 2: Store these values into a 250 samples sized window (1s) so that
            1 window corresponds to a row of dimensions (nb_features * nb_channels)
    Step 3: Slide the window by 0.1s/25 samples and repeat the process
    Step 4: Return a matrix of dimension (nb_window, nb_features*nb_channels)
    '''
    
    # Empty matrix with columns corresponding to correspond to 10 features * 20 channels = 200
    # Each row will correspond to 1 window.
    wins = np.zeros((((eeg_chans.shape[1]//sld_size)-win_size + sld_size),\
                        int((fs/feat_size) * eeg_chans.shape[0])))
    
    for k in range(len(wins)):# length of continuous data / slide size until last window.
    
        for j in range(eeg_chans.shape[0]):# Number of channels
    
            for i in range(nb_feat):
        
                wins[k, i + 10 * j] = np.mean(eeg_chans.iloc[j, (0 + feat_size * i) + k * sld_size: \
                                                             (feat_size + feat_size * i) + k * sld_size])
        
    return wins
```

<font size="3">    <b>Application of previously trained classifier and get the probabilities of movement intent</b>

```python
def prediction(wins):
    '''
    Step 1: Apply the classifier on each window
    Step 2: Extract the probability value corresponding to class motion
    Step 3: Interpolate the probabilities and make the array from n_windows to the length of the continuous data ?
    '''
    prob = np.zeros((wins.shape[0]))
    
    for i in range(wins.shape[0]):
        
        prob[i] = clf_eeg.predict_proba(wins[i,:].reshape(1, -1))[0][1]
        
    return prob
```
